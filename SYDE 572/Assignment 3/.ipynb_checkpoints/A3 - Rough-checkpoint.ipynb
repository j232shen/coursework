### Exercise 1: Non-Parametric Estimation

```python
import torch
import torchvision
import torchvision.datasets as datasets

print("PyTorch version:", torch.__version__)
```

```python
# initializing training set
mnist_trainset = datasets.MNIST(root='./data', train=True, download=True, transform=None)
mnist_train_loader = torch.utils.data.DataLoader(mnist_trainset, batch_size=64, shuffle=True)

# initializing test set
mnist_testset = datasets.MNIST(root='./data', train=False, download=True, transform=None)
mnist_test_loader = torch.utils.data.DataLoader(mnist_testset, batch_size=64, shuffle=True)
```

```python
# verify the original shape of the images for the training set
original_train_shape = mnist_trainset.data.shape
print("Original shape of training images:", original_train_shape)

# verify the original shape of the images for the test set
original_test_shape = mnist_testset.data.shape
print("Original shape of test images:", original_test_shape)
```

```python
# flatten the training set
flattened_train_images = mnist_trainset.data.view(mnist_trainset.data.size(0), -1)
flattened_train_labels = mnist_trainset.targets

# flatten the test set
flattened_test_images = mnist_testset.data.view(mnist_testset.data.size(0), -1)
flattened_test_labels = mnist_testset.targets
```

```python
# verify the shape of flattened images for the training set
print("Shape of flattened training images:", flattened_train_images.shape)

# verify the shape of flattened images for the test set
print("Shape of flattened test images:", flattened_test_images.shape)
```

```python
from sklearn.decomposition import PCA

# create a PCA object with 1 component
pca = PCA(n_components=1)

# fit the PCA model to the flattened training images
pca.fit(flattened_train_images)

# transform the training set and test set images into 1x1 vectors using PCA
transformed_train_images = pca.transform(flattened_train_images)
transformed_test_images = pca.transform(flattened_test_images)
```

```python
# verify the shape of the transformed training images
print("Shape of transformed training images:", transformed_train_images.shape)

# verify the shape of the transformed test images
print("Shape of transformed test images:", transformed_test_images.shape)
```

```python
# filter the training set to include only classes 0 and 1
indices_train = (flattened_train_labels == 0) | (flattened_train_labels == 1)
filtered_train_images = transformed_train_images[indices_train]
filtered_train_labels = flattened_train_labels[indices_train]

# filter the test set to include only classes 0 and 1
indices_test = (flattened_test_labels == 0) | (flattened_test_labels == 1)
filtered_test_images = transformed_test_images[indices_test]
filtered_test_labels = flattened_test_labels[indices_test]
```

#### Question 1: Histogram-based estimation

```python
# find the minimum and maximum values in the dataset
min_value = np.min(filtered_train_images)
max_value = np.max(filtered_train_images)

print("Minimum value in the dataset:", min_value)
print("Maximum value in the dataset:", max_value)
```

```python
ranges = [1, 10, 100]
labels = [0, 1]

for label in labels:
    # Filter data for the current class
    class_data = filtered_train_images[filtered_train_labels == label]
    
    for r in ranges:
        num_bins = round((max_value - min_value)/r)
        
        # calculate histogram for the current class and range
        hist, bin_edges = np.histogram(class_data, bins=num_bins, density=True)
        bin_widths = bin_edges[1:] - bin_edges[:-1]

        plt.bar(bin_edges[:-1], hist, width=bin_widths, align='edge')
        plt.xlabel("Values")
        plt.ylabel("Probability")
        plt.title(f"Class {label} - Estimated Probability Distribution (Range: {r})")
        plt.show()
```

#### Question 2: ML Classifier

```python
def ML_classify(width, x):
    num_bins = round((max_value - min_value)/width)
    
    # define the bin edges explicitly to cover the full data range
    bin_edges = np.linspace(min_value, max_value, num=num_bins+1, endpoint=True)
    
    # filtering for classes
    c1 = filtered_train_images[filtered_train_labels == 0]
    c2 = filtered_train_images[filtered_train_labels == 1]
    
    # calculate histogram for c1
    hist1, _ = np.histogram(c1, bins=bin_edges, density=True)
    
    # calculate histogram for c2
    hist2, _ = np.histogram(c2, bins=bin_edges, density=True)
    
    # determine which bin x belongs to in both histograms
    if x >= max_value:
        bin_idx_c1 = len(hist1)-1
        bin_idx_c2 = len(hist2)-1
    else:
        bin_idx_c1 = np.digitize(x, bin_edges) - 1  # subtract 1 to convert to 0-based indexing
        bin_idx_c2 = np.digitize(x, bin_edges) - 1

    # access the heights of the bins for x in both histograms
    height_c1 = hist1[bin_idx_c1]
    height_c2 = hist2[bin_idx_c2]

    # compare the heights to classify
    if height_c1 > height_c2:
        return 0
    else:
        return 1
```

```python
ranges = [1, 10, 100]

for r in ranges:
    print(f"Range = {r}")
    predicted_labels = []
    for sample in filtered_test_images:
        predicted_labels.append(ML_classify(r, sample))
        
    # calculate classification accuracy
    accuracy = np.mean(filtered_test_labels.numpy() == predicted_labels)
    print(f"Accuracy for range = {r}: {accuracy * 100:.2f}%\n")
```

#### Question 3: Kernel-based density estimation

```python
def gaussian_kernel(x, mean, sigma):
    return (1 / (sigma * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - mean) / sigma) ** 2)

def kd_classify(x, mean1, mean2, sigma):
    # calculate probability densities for both classes
    c1_density = gaussian_kernel(x, mean1, sigma)
    c2_density = gaussian_kernel(x, mean2, sigma)
    
    # compare densities to classify
    if c1_density > c2_density:
        return 0
    else:
        return 1
```

```python
c1 = filtered_train_images[filtered_train_labels == 0]
c2 = filtered_train_images[filtered_train_labels == 1]

c1_mean = np.mean(c1)
c2_mean = np.mean(c2)
sigma = 20

predicted_labels = []
for sample in filtered_test_images:
    predicted_labels.append(kd_classify(sample, c1_mean, c2_mean, sigma))

# calculate classification accuracy
accuracy = np.mean(filtered_test_labels.numpy() == predicted_labels)
print(f"Accuracy = {accuracy * 100:.2f}%")
```

### Exercise 2: K-means Clustering

```python
# initializing training set
mnist_trainset = datasets.MNIST(root='./data', train=True, download=True, transform=None)
mnist_train_loader = torch.utils.data.DataLoader(mnist_trainset, batch_size=64, shuffle=True)

# initializing test set
mnist_testset = datasets.MNIST(root='./data', train=False, download=True, transform=None)
mnist_test_loader = torch.utils.data.DataLoader(mnist_testset, batch_size=64, shuffle=True)

# flatten the training set
flattened_train_images = mnist_trainset.data.view(mnist_trainset.data.size(0), -1)
flattened_train_labels = mnist_trainset.targets

# flatten the test set
flattened_test_images = mnist_testset.data.view(mnist_testset.data.size(0), -1)
flattened_test_labels = mnist_testset.targets

# rename for clarity
train_data = flattened_train_images.numpy()
train_labels = flattened_train_labels.numpy()
test_data = flattened_test_images.numpy()
test_labels = flattened_test_labels.numpy()

# verify the shape of flattened images for the training set
print("Shape of flattened training images:", train_data.shape)

# verify the shape of flattened images for the test set
print("Shape of flattened test images:", test_data.shape)
```

#### Question 1: K-means algorithm implementation

```python
class Kmeans:
    def __init__(self, x, k):
        self.x = x
        self.k = k
        self.prototypes = None
    
    def fit(self, tolerance=1e-4, max_iterations=100):
        # random initialization of prototypes
        np.random.seed(42)
        indices = np.random.choice(self.x.shape[0], self.k, replace=False)
        self.prototypes = self.x[indices]

        # repeat until convergence
        for _ in range(max_iterations):
            labels = []
            for sample in self.x:
                # calculate euclidean distances between sample and all prototypes
                distances = []
                for prototype in self.prototypes:
                    distances.append(np.linalg.norm(prototype - sample))

                # cluster label is index of minimum sample-prototype distance
                cluster = distances.index(min(distances))
                labels.append(cluster)

            # initializing empty list of cluster means
            cluster_means = []

            # iterating through clusters
            for cluster in range(self.k):
                # find all indices where label == cluster
                indices = [idx for idx, label in enumerate(labels) if label == cluster]

                # calculate mean of cluster
                cluster_mean = np.mean(self.x[indices], axis=0, keepdims=True)
                cluster_means.append(cluster_mean)
                
            # check for convergence
            new_prototypes = np.vstack(cluster_means)
            if np.all(np.isclose(new_prototypes, self.prototypes, rtol=tolerance)):
                break # converged            

            # set cluster means as new prototypes
            self.prototypes = cluster_means
        
    def predict(self, test_data):
        labels = []
        for sample in test_data:
            # calculate euclidean distances between sample and all prototypes
            distances = []
            for prototype in self.prototypes:
                distances.append(np.linalg.norm(prototype - sample))

            # cluster label is index of minimum sample-prototype distance
            cluster = distances.index(min(distances))
            labels.append(cluster)
        
        return labels
```

#### Question 2

```python
import time

k_values = [5, 10, 20, 40]
cluster_labels = {'5': [], '10': [], '20': [], '40': []}

for k in k_values:
    print(f"k = {k}")
    start = time.time()
    
    kmeans = Kmeans(train_data, k)
    kmeans.fit()
    cluster_labels[str(k)] = kmeans.predict(test_data)
    
    end = time.time()
    print(f"Elapsed time: {(end - start):.3f}s\n")
```

```python
consistency_scores = []

for k in k_values:
    print(f"k = {k}")
    Q_values = []
    
    # iterating through each cluster
    for i in range(k):
        cluster_indices = np.where(np.array(cluster_labels[str(k)]) == i)[0]
        cluster_data = test_labels[cluster_indices]

        class_counts = np.bincount(cluster_data)
        max_count = class_counts.max()
        Q = max_count / len(cluster_data)
        Q_values.append(Q)
        
    Q_total = np.mean(Q_values)
    print(f"Q_total = {Q_total}\n")
    consistency_scores.append(Q_total)
```

### Exercise 3: Gaussian Mixture Model (GMM)

#### Question 1

```python
import numpy as np


def calculate_responsibility(x, k, pi, mu, sigma):
    """
    Calculate the responsibility for a data point xi and a Gaussian component k in a GMM.

    Parameters:
    - x: Data point (a 1D NumPy array).
    - k: Index of the Gaussian component.
    - pi: Mixture weights for all components (a 1D NumPy array).
    - mu: Means for all components (a 2D NumPy array).
    - sigma: Covariance matrices for all components (a list of 2D NumPy arrays).

    Returns:
    - r_k: Responsibility for x and component k.
    """
    d = len(x)  # Dimension of data

    # Calculate the likelihood of x under the Gaussian component k
    diff = x - mu[k]
    sigma_inv = 1.0 / sigma[k] # inverse of diagonal cov matrix
    exponent = -0.5 * (diff @ (sigma_inv @ diff))
    likelihood = np.exp(exponent) / np.sqrt(np.prod(sigma[k])) # det of diagonal = product of diagonal elements

    # Calculate the numerator of the responsibility formula
    numerator = pi[k] * likelihood

    # Calculate the denominator of the responsibility formula
    denominator = 0.0
    for j in range(len(pi)):
        diff_j = xi - mu[j] 
        sigma_inv_j = 1.0 / sigma[j] # inverse of diagonal cov matrix
        exponent_j = -0.5 * (diff_j @ (sigma_inv_j @diff_j))
        likelihood_j = np.exp(exponent_j) / np.sqrt(np.prod(sigma[j])) # det of diagonal = product of diagonal elements
        denominator += pi[j] * likelihood_j

    # Calculate the responsibility
    r_k = numerator / denominator
    
    return r_k


def gmm_em(X, K, max_iter=500, tol=1e-6):
    n, d = X.shape  # Number of data points and dimensions

    # Initialize GMM parameters
    pi = np.ones(K) / K  # Initialize mixture weights uniformly
    mu = X[np.random.choice(n, K, replace=False)]  # Randomly initialize means
    sigma = [np.ones(d) for _ in range(K)]  # Initialize diagonal covariance vectors with ones

    for iteration in range(max_iter):
        # E-step: Calculate responsibilities using the custom function
        r_ik = np.zeros((n, K))
        for i in range(n):
            for k in range(K):
                r_ik[i, k] = calculate_responsibility(X[i], k, pi, mu, sigma)
        
        # Calculate log-likelihood
        log_likelihood = np.sum(np.log(np.sum(r_ik, axis=1)))

        if iteration > 0 and np.abs(log_likelihood - prev_log_likelihood) < tol * np.abs(log_likelihood):
            break
                                            
        # M-step: Update parameters
        r_ik_sum = np.sum(r_ik, axis=0)
        pi = r_ik_sum / n
        mu = (r_ik.T @ X) / r_ik_sum[:, np.newaxis]
        
        for k in range(K):
            diff = X - mu[k]
            sigma[k] = np.sum((r_ik[:, k] * (diff ** 2)), axis=0)

        prev_log_likelihood = log_likelihood

    return pi, mu, sigma, log_likelihood
```

Responsibility Calculation:
"In the Gaussian Mixture Model (GMM) EM algorithm with diagonal covariance matrices, the responsibility rik is derived to determine the likelihood that a data point xi belongs to Gaussian component k. The process begins with the evaluation of the likelihood of xi under component k, which considers the multivariate Gaussian probability density function. This likelihood is calculated with respect to the component's mean µk and diagonal covariance vector sigma[k]. The likelihood is then weighted by the mixture weight πk assigned to component k. The denominator for the responsibility is established by summing the weighted likelihoods across all components, accounting for their respective πk. The responsibility rikis subsequently determined as the ratio of the weighted likelihood for componentk to the denominator. This calculation provides a measure of the likelihood that a data point is associated with a particular component, and it is a key step in the EM algorithm's expectation step."

Update of sigma[k] (Diagonal Covariance Vector):
"In the Maximization (M) step of the EM algorithm, the diagonal covariance vector sigma[k] for each component is updated to better fit the data. This update involves the calculation of the weighted sum of squared differences between data points xi and the mean µk of component k. This sum is carried out for all data points, generating a vector with the same dimension as the data. To obtain the updated sigma[k], the result from this sum is divided by the sum of responsibilities associated with component k. The updated sigma[k] is thus tailored to reflect the weighted average of the squared differences between data points and the component's mean. This process ensures that the diagonal covariance vectors adapt to the data points that are assigned to the respective components and plays a crucial role in parameter estimation during the M-step of the EM algorithm."
